Taking Squirrel Data and Investigating

```python
# First import data #
import pandas as pd
import os
data_set = pd.read_csv('/content/squirrel-data.csv', encoding='cp1252')
data_set
```

```python
# Remove empty data points #

data_set_edited = data_set[data_set["Squirrel Latitude (DD.DDDDDD)"].notna()]
data_set_edited = data_set_edited[data_set["Squirrel Longitude (-DD.DDDDDD)"].notna()]
data_set_edited
```

```python
import folium
```

```python
m=folium.Map(location=[40.841359,-73.934339], min_zoom = 11, max_zoom = 25)
m
```

```python
# Place markers onto the map from the data set #
for each in range(len(data_set_edited)):

    folium.Marker([data_set_edited["Squirrel Latitude (DD.DDDDDD)"].iloc[each],data_set_edited["Squirrel Longitude (-DD.DDDDDD)"].iloc[each]], popup = data_set_edited["Squirrel ID"].iloc[each]).add_to(m)
m
```
